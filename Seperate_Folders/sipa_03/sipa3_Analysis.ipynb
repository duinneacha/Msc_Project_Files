```python
import cv2
import matplotlib.pyplot as plt

# List of image filenames
image_filenames = ['./sipaimages/2/22423_20220321161435.jpg', './sipaimages/2/22420_20220321135314.jpg', './sipaimages/2/8.jpg', './sipaimages/2/42533_20220505081840.jpg', './sipaimages/2/23039_20220326173036.jpg']



# List to store the contrasts of the images
contrasts = []

# Loop through the list of image filenames
for filename in image_filenames:
    # Load the image using OpenCV
    image = cv2.imread(filename)

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Compute the contrast of the image by calculating the standard deviation of the pixel intensities
    contrast = cv2.meanStdDev(gray_image)[1][0][0]

    # Add the contrast value to the list
    contrasts.append(contrast)

# Plot the contrast values using Matplotlib
plt.plot(contrasts, 'ro', markersize=3)

# Add the file names as labels for the points
for i, filename in enumerate(image_filenames):
    plt.text(i, contrasts[i], filename, fontsize=5)

plt.show()
```
