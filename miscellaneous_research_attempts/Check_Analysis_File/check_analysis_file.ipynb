```python
import pandas as pd
```

```python
df = pd.read_excel(r'Images_Tesseract_Analysis.xlsx')
df = df.reset_index()
df
```

```python
def check_index(index, index_array):
    if index not in index_array:
        index_array.append(index)
    
    
    
```

```python
folder_dic = {}

index_array = []
t_sevsega_eng = 0
t_sevsega_ssd = 0
t_sevseg_contours_ssd = 0
t_sevseg_contours_eng = 0
t_thresh_ssd = 0
t_thresh_eng = 0
t_closing_ssd = 0
t_closing_eng = 0
t_otsu_ssd = 0
t_otsu_eng = 0

for index, row in df.iterrows():
    if row['seen_value']==row['sevsega_eng']:
        t_sevsega_eng += 1
        check_index(index, index_array)
        continue
        
        
    if row['seen_value']==row['sevsega_ssd']:
        t_sevsega_ssd += 1
        check_index(index, index_array)
        continue

    if row['seen_value']==row['sevseg_coutours_ssd']:
        t_sevseg_contours_ssd += 1
        check_index(index, index_array)
        continue

    if row['seen_value']==row['sevseg_coutours_eng']:
        t_sevseg_contours_eng += 1
        check_index(index, index_array)
        continue

    if row['seen_value']==row['thresh_ssd']:
        t_thresh_ssd += 1
        check_index(index, index_array)
        continue

    if row['seen_value']==row['thresh_eng']:
        t_thresh_eng += 1
        check_index(index, index_array)
        continue

    if row['seen_value']==row['closing_ssd']:
        t_closing_ssd += 1
        check_index(index, index_array)
        continue

    if row['seen_value']==row['closing_eng']:
        t_closing_eng += 1
        check_index(index, index_array)
        continue

    if row['seen_value']==row['otsu_ssd']:
        t_otsu_ssd += 1
        check_index(index, index_array)
        continue

    if row['seen_value']==row['otsu_eng']:
        t_otsu_eng += 1
        check_index(index, index_array)
        print(index)
        continue
        



print("sevsega:", "eng", t_sevsega_eng, "ssd", t_sevsega_ssd)
print("sevdeg_contours:", "eng", t_sevseg_contours_eng, "ssd", t_sevseg_contours_ssd)
print("thresh:", "eng", t_thresh_eng, "ssd", t_thresh_ssd)
print("closing:", "eng", t_closing_eng, "ssd", t_closing_ssd)
print("otsu:", "eng", t_otsu_eng, "ssd", t_otsu_ssd)

len(index_array)
```
