```python
import cv2
import numpy as  np
import glob
from matplotlib import pyplot as plt
```

```python
# Function to show an image using pyplot
def show_img(img, size=12, cmap=None):
  fig = plt.gcf()
  fig.set_size_inches(size, size)
  plt.axis("off")
  # plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
  plt.imshow(img, cmap)
  plt.show()
```

```python
def display(img, cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
img1 = cv2.imread("./sipaimages/test/13.jpg")
show_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread("./sipaimages/test/22419_20220321130145.jpg")
show_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img3 = cv2.imread("./sipaimages/test/27.jpg")
show_img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

img4 = cv2.imread("./sipaimages/test/43359_20220517131209.jpg")
show_img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)

show_img(img1, 3, cmap="gray")
show_img(img2, 3)
show_img(img3, 3)
show_img(img4, 3)
```

```python
hist_values = cv2.calcHist([img2], channels=[0], mask=None, histSize=[256], ranges = [0,256])
plt.plot(hist_values)
```

```python
img = img4

color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.cv2.calcHist([img], channels=[i], mask=None, histSize=[256], ranges=[0,256])
    plt.plot(histr, color = col)
    plt.xlim(0, 256)
    plt.ylim(0, 15000)

plt.title("Histogram for img")
```
