## Contrast in OpenCV

In image processing, contrast refers to the difference in intensity between the light and dark areas of an image. High contrast means that there is a large difference in intensity between the light and dark areas, while low contrast means that the difference in intensity is small.

OpenCV is an open-source computer vision library that provides tools for image and video processing. One of the operations that OpenCV provides is the ability to adjust the contrast of an image. This is often done to enhance the visual appearance of an image or to improve its visibility.

To adjust the contrast of an image in OpenCV, you can use the cv2.addWeighted() function. This function takes in three parameters: the input image, a weight for the image, and a bias value. The weight and bias values can be used to adjust the contrast of the image. A larger weight value will increase the contrast, while a smaller weight value will decrease the contrast. The bias value can be used to shift the intensity values of the image up or down, which can also affect the overall contrast of the image.

Here is an example of how to use the cv2.addWeighted() function to increase the contrast of an image in OpenCV:


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Function to show an image using pyplot
def show_img(img, size=12):
  fig = plt.gcf()
  fig.set_size_inches(size, size)
  plt.axis("off")
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
  plt.show()
```

```python
img = cv2.imread('./sipaimages/2/22420_20220321135314.jpg')
show_img(img, 5)
```

```python
alpha = 1.5
beta = 50

result = cv2.addWeighted(img, alpha, np.zeros(img.shape, img.dtype), 0, beta)

show_img(result, 5)
```

```python
import cv2
import pytesseract

# Read the image and convert it to grayscale
image = cv2.imread('./sipaimages/2/22420_20220321135314.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Adjust the contrast of the image
gray = cv2.equalizeHist(gray)

# Remove noise from the image using median filtering
gray = cv2.medianBlur(gray, 3)

# Binarize the image using adaptive thresholding
thresh = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# Pass the image to Tesseract OCR and extract the text
text = pytesseract.image_to_string(thresh)
print(text)
```

```python
show_img(gray,5)
show_img(thresh,5)
```
