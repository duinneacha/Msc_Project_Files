```python
from PIL import Image
import pytesseract
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Function to show an image using pyplot
def show_img(img, size=12):
  fig = plt.gcf()
  fig.set_size_inches(size, size)
  plt.axis("off")
  # plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
  plt.imshow(img)
  plt.show()
```

```python
import numpy as np
import cv2

img = cv2.imread('images/22430_20220322115422.jpg')

norm_img = np.zeros((img.shape[0], img.shape[1]))
img2 = cv2.normalize(img, norm_img, 0, 255, cv2.NORM_MINMAX)
img3 = cv2.threshold(img2, 100, 255, cv2.THRESH_BINARY)[1]
img4 = cv2.GaussianBlur(img3, (1, 1), 0)

show_img(img,5)
show_img(img2,5)
show_img(img3,5)
show_img(img4,5)
img2
```
