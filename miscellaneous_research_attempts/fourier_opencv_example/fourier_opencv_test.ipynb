```python
import cv2
import numpy as  np
import glob
from matplotlib import pyplot as plt
```

```python
# Function to show an image using pyplot
def show_img(img, size=12):
  fig = plt.gcf()
  fig.set_size_inches(size, size)
  plt.axis("off")
  # plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
  plt.imshow(img)
  plt.show()
```

```python
# list_images = glob.iglob("./letters/*")
list_images = glob.iglob("./sipaimages/test/*")

```

```python
for image_title in list_images:
    print(image_title)
    img = cv2.imread(image_title, cv2.IMREAD_GRAYSCALE)
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    magnitude_spectrum = 20*np.log(np.abs(fshift))
    magnitude_spectrum = np.asarray(magnitude_spectrum, dtype=np.uint8)
    img_and_magnitude = np.concatenate((img, magnitude_spectrum), axis=1)

    show_img(img_and_magnitude, 5)
```
