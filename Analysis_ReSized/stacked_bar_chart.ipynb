```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ["USA", "Russia", "GBR", "China", "India"]

gold = [20, 10, 6, 24, 30]
silver = [100, 200, 30, 120, 150]
bronze = [200, 120, 60, 240, 300]


b_bronze = list(np.add(gold, silver))


plt.bar(x, gold, 0.4, label="Gold")
plt.bar(x, silver, 0.4, bottom=gold, label="Silver")
plt.bar(x, bronze, 0.4, bottom=b_bronze, label="Bronze")
# plt.bar(x, bronze, 0.4, bottom=silver, label="Bronze")

plt.xlabel("Countries")
plt.ylabel("Medals")
plt.title("Medals vs Countries")
plt.legend()
plt.show()


```

```python
w = 0.4

x = ["CSE", "ISE", "ECE", "EEE", "Mech"]

boys = [100, 40, 60, 240, 300]
girls = [200, 70, 30, 120, 15]

plt.bar(x, boys, w, label="Boys", color="lightblue")
plt.bar(x, girls, w, bottom = boys, label="Girls", color="pink")

plt.xlabel("Courses")
plt.ylabel("Students")
plt.title("Students Vs Courses")
plt.legend()
plt.show()
```
