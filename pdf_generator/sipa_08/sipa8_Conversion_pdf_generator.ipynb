```python




# Read the image
image_path = "./cropped_images/8/f01_cropped.png"
image = cv2.imread(image_path)
adfns.show_img(image, size=3)
# Convert to HSV and apply color filtering
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

lower_green = np.array([40, 40, 40])
upper_green = np.array([90, 255, 255])

mask = cv2.inRange(hsv_image, lower_green, upper_green)


masked_green = adfns.mask_green(image)
masked_green_gray = cv2.cvtColor(masked_green, cv2.COLOR_BGR2GRAY)
adfns.show_img(masked_green, size=3, title="masked_green")
adfns.show_img(masked_green_gray, size=3, title="masked_green_gray")

# Adaptive thresholding
thresh_img = cv2.adaptiveThreshold(masked_green_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 11, 2)

# Morphological operations
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
dilated_img = cv2.dilate(thresh_img, kernel, iterations=1)

# Remove noise
contours, _ = cv2.findContours(dilated_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
for contour in contours:
    if cv2.contourArea(contour) < 100:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(dilated_img, (x, y), (x + w, y + h), 0, -1)

# Invert image (if necessary)
inverted_img = cv2.bitwise_not(dilated_img)
adfns.show_img(inverted_img, size=3)
# Pass the preprocessed image to pytesseract
result = pytesseract.image_to_string(inverted_img)

print("result:", result)

```
